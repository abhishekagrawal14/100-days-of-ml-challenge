```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[: , 0] , X[: , 1] , c=y , cmap = 'winter' , s= 100)
plt.show()
```

```python
def step(z):
    if z >= 0 :
        return 1 
    else :
        return 0
```

The main perceptron function 

```python
def perceptron(X,y):
    X = np.insert(X,0,1,axis =1)
    weights = np.ones(X.shape[1])
    # weights is a matrix -> weights = [w0,w1,w2] just three weights here 
    # x is also a matrix it is a row matrix while w is column matrix .
    lr = 0.1
    epochs = 1000 

    for i in range(epochs):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]

    return weights[0] ,weights[1:]
```

```python
intercept_,coeff_ = perceptron(X,y)
```

```python
print(intercept_)
print(coeff_)
```

```python
m = -(coeff_[0]/coeff_[1])
c = -(intercept_/coeff_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input+c
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
