```python
import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

here ast.literal_eval ek function  inside ast library jo is strings value ko sach me list of dictionary bana de rha hai phir usse hamkon bas names chahiye id hamare liye junk hai tf-idf me noise hoga usse .

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
            
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break   # ek hi director milega generally, mil gaya toh ruk jao
    return L
```

```python
df['genres'] = df['genres'].apply(convert)
df['keywords']= df['keywords'].apply(convert)
df['cast']=df['cast'].apply(convert3)
df['crew']= df['crew'].apply(fetch_director)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

new_df['tags'] = new_df['tags'].apply(stem)
```

stemming -> basically humlog sabhi words ko unke root stem me kar rhe jaise lovely , loved , loving all will convert to love 

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=5000, stop_words='english')
vectors = tfidf.fit_transform(new_df['tags']).toarray()
```

tf dif wala step yaha humlog stop_words = english karke , a , an , the , he , she , it jaise kachre hata rahe hai , tf dif pure data ko vector me convert kar diya 

```python
vectors
```

cosine similarity ye 2 vectors ke bich me simmilarities check karta hai .

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:11]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```

```python
recommend( "The Avengers" )
```
